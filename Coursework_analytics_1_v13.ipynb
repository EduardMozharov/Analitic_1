#### 1.3.6 Минимальный, максимальный, средний, медианный возраст студентов

```python
# Поработаем с учениками
students.head()
```

```python
# Создадим датафрейм со студентами у которых есть birthday
student_age = students[students['birthday'].notna()] 

student_age.head()
```

```python
student_age['birthday_dt'] = pd.to_datetime(student_age['birthday'], errors='coerce')
#student_age['birthday_dt'] = student_age['birthday_dt'].to_timestamp
student_age.head()                                                                             
```

```python
# Хочется избежать перебора циклом
#student_age['Age'] = (pd.to_datetime("today").year - student_age['birthday_dt'].year)  +\   
                    # int(pd.to_datetime("today").dayofyear > student_age['birthday_dt'].dayofyear)
# но после длительных попыток вычесть даты используя pd.Timestamp
# не получилось
    
today_year = pd.to_datetime("today").year           # Теккущий год
today_dayofyear = pd.to_datetime("today").dayofyear # Текущий день от начала года
N = len(student_age['birthday_dt'])
student_age['Age'] = 0

for i in range(N):
    student_age['Age'].iloc[i] = today_year - student_age['birthday_dt'].iloc[i].year # Разница в календалных годах
    
    # Если год прошел не полный, то возраст меньше на 1
    if today_dayofyear < student_age['birthday_dt'].iloc[i].dayofyear:
        student_age['Age'].iloc[i] -= 1 
    
student_age.head()   
```

```python
#student_age = student_age[student_age['Age'] > 0]
print('6) Возраст студентов:')
print('Минимальный:', student_age['Age'].min())
print('Максимальный:', student_age['Age'].max())
print('Средний:', student_age['Age'].mean())
print('Медианный:', student_age['Age'].median())
```

#### 1.3.9 Количество студентов на каждом курсе

```python
sphere = main['field'].unique() # Сферы деятельности
sphere_color =  pd.DataFrame([['r'], ['g'], ['b'], ['m']], index=sphere)
sphere_color
```

```python
students_count_sorted = students_count.sort_values('N', ascending=False)

fig, ax = plt.subplots(figsize=(12, 6))
short_legend = []

for course_name in students_count_sorted.index: 
    field_name = main[main['title'] == course_name]['field'].iloc[0]
    
    # Горизонтальные столбцы, Добавление имен графиков для легенды, прозрачность 0.9
    # C прозрачностью 0.1 ничего не видно, заменил на 0.9
    # Формировние легенды из четырех элементов
    if field_name in short_legend:
        ax.barh(course_name, students_count_sorted['N'].loc[course_name], \
            alpha=0.9,
            color = sphere_color.loc[field_name])  
    else:
        ax.barh(course_name, students_count_sorted['N'].loc[course_name], \
            alpha=0.9,
            color = sphere_color.loc[field_name],\
            label = field_name)     
        short_legend.append(field_name)

ax.set_title("Количество студентов на каждом курсе")
ax.set_ylabel('Наименование курса')
ax.set_xlabel('Количество студентов')
ax.legend()    # Отображение легенды

# Отображение медианы
st_median = students_count_sorted.median()
ax.axvline(int(st_median), color='y')

# Удаляем рамку
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)

plt.show()

# График должен иметь заголовок. Значения должны быть отсортированы. Цвет столбцов должен содержать информацию о сфере, к которой относится курс (то есть нужна легенда). Прозрачность должна стоять на отметке 0.1. На график должна быть нанесена линия медианы. У медианы должен быть свой цвет. Рамки у графика быть не должно ⭐
# На основании рассчитанных значений опишите данные (описание должно быть полным и покрывать все полученные выше метрики)
```

```python
# Введем функцию,которая позволит найти все месяцы между началом и окончанием выполнения ДЗ
# Функция вычитания двух дат с записью всех лет и месяцев в них входящих
def find_count_months(start_year, start_month, finish_year, finish_month):
    y_m = [[start_year, start_month]]
    while (finish_year != start_year) or (finish_month != start_month):
        # Если года неравны
        if finish_year != start_year:
            # Если декабрь, то перешагиваем год
            if start_month == 12:
                start_month = 1
                start_year  += 1
            else:
                start_month += 1
          
        # Если года равны
        else:
            start_month += 1
        y_m.append([start_year, start_month])
    return y_m
```

## 4.1  Посчет конверсии перехода студентов из одного модуля в другой на каждом курсе
**(p)** Посчитать конверсию перехода студентов из одного модуля в другой на каждом курсе. Формула: отношение количества студентов, приступивших к выполнению домашнего задания в этом модуле (если дз в модуле несколько, то считать по первому дз в модуле), к количеству студентов, сдавших задание в предыдущем модуле (если дз в модуле несколько, то считать по последнему дз в модуле).


```python
# Загрузим данные (что бы не повторять пересчеты в первой и второй частей)
main = pd.read_csv('main.csv')
del main['Unnamed: 0']
main.head()
```

```python
main['status'].unique()
```

```python
# Найдем количество студентов выполнивших или приступивших к первому ДЗ модуля
df_module_1hw_start = main[ (main['is_homework']) & \
                            (main['status'].isin(['done', 'start', 'fail', 'wait']))] \
                      .groupby(['title', 'module_number', 'lesson_number']).count()\
                      .reset_index()\
                      .groupby(['title', 'module_number']).first()\
                      [['lesson_number', 'progress_id']]\
                      .rename(columns={'progress_id': 'N_students_1hw_start'})

df_module_1hw_start.head()
```

```python
# Найдем количество студентов выполнивших последнее ДЗ в модуле
df_module_last_hw_done = main[ (main['is_homework']) & \
                               (main['status'].isin(['done']))] \
                         .groupby(['title', 'module_number', 'lesson_number']).count()\
                         .reset_index()\
                         .groupby(['title', 'module_number']).last()\
                         [['lesson_number', 'progress_id']]\
                         .rename(columns={'progress_id': 'N_students_last_hw_done'})

df_module_last_hw_done.head()
```

```python
## Объединим 
df = df_module_1hw_start.reset_index()\
     .merge(df_module_last_hw_done.reset_index())
df
```

```python
# Подсчитаем конверсию
titles = df.title.unique()
conversia = []
conversia_drob = []
for title in titles:
    df_title = df[df['title'] == title]
    for i in range(df_title.shape[0]):
        if i == 0:
            val = 0
            drob = 0
        else:
            val = df_title.iloc[i][3] / df_title.iloc[i-1][4]
            drob = str(df_title.iloc[i][3]) +' / ' + str(df_title.iloc[i-1][4])
        conversia.append(val)
        conversia_drob.append(drob)
        
print(len(conversia))  
```

```python
# Добавим столбец конверсии
df['Conversion'] = conversia
df['Conversion_drob'] = conversia_drob
df.head()
```

## 4.2 График, отражающий конверсию перехода студентов из одного модуля в другой на каждом курсе
(m) Постройте bar-chart, отражающий конверсию перехода студентов из одного модуля в другой на каждом курсе. График должен иметь заголовок. Ticks нужно развернуть так, чтобы они были читаемы

```python
fig, ax = plt.subplots(figsize=(16, 8))

index = 1
bar_width = 0.05

for title in titles:
    df_title = df[df['title'] == title]
    for i in range(1, df_title.shape[0]):
        plt.bar(index+bar_width*i, df_title['Conversion'].iloc[i], bar_width,\
            alpha=0.8)
    index +=1
    
ax.set_title('Конверсия перехода студентов из одного модуля в другой на каждом курсе', fontsize=20)
ax.set_xlabel('Курс', fontsize=14)
ax.set_ylabel('Конверсия', fontsize=14)
#ax.set_yticks(range(0, 2200, 200))
ax.set_xticks( range(1, index))
ax.set_xticklabels(titles.tolist(), rotation=80)
plt.grid()

plt.show()
```

```python
sphere = main['field'].unique().tolist()

df_sphere_color = main[[ 'module_number', 'module_title', 'title' ,'field']]
df_sphere_color['color'] =  [ ['r', 'g', 'b', 'm'][sphere.index(elem)] for elem in df_sphere_color['field'] ]

df_sphere_color.head()
```

```python
df['N._module_title'] = df['module_number'].astype(str).str.cat(df['module_title'], sep =". ")
```
